# Demonstration of orbweaver network simulation tools

```python
%matplotlib inline
from orbweaver.metrics import Metrics
from orbweaver.dataset import Dataset
from orbweaver.network import Network, export_graph
from orbweaver.visuals import Visuals
from orbweaver.simulate import Simulator
from orbweaver.profiles import Profile
```

```python
test = Simulator.random_dataset(length=10, size=10, a=-1, b=1, name="random_data")
test.data
```

```python
test.data.any() > 0
```

```python
R = Network()
R.build_matrix(test, metric=Metrics.cos_sim, org="columns", save=False)
Visuals.heatmap(R.matrix)
```

```python
R.build_graph("random_network")
Visuals.quickdraw(R.graph,  weights=R.stats['weights'])
```

```python
R.quickstats()
Visuals.outlier_plot(R)
```

```python
struct = Simulator.structured_dataset(seeds=0.1, profiles = {'linear': 1/3, 'powerlaw': 1/3, 'multiplicative': 1/3}, 
                           size=10, free=0.1, length=50, name="structured_simulation")
struct.data
```

```python
Rs = Network()
Rs.build_matrix(struct, metric=Metrics.cos_sim, org="columns", save=False)
Visuals.heatmap(Rs.matrix)
```
